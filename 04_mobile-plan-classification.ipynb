Introduction: In this project we are using several models to assist a mobile carrier company in suggesting/predicting which plan subscribers would switch to. Aiming for a threshold of accuracy 0.75, and also checking using testing and a sanity check, we find the best model to use for this particular goal. Below, I use the models: random forest, decision tree, and logistic regression, to train and test on a percentage of the data. I later come to the conclusion that one model stands out for this goal, and explain why.

````python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

df = pd.read_csv('/datasets/users_behavior.csv')

print(df) # calls  minutes  messages   mb_used  is_ultra
print(df.head())
print(df.info())

features = df.drop('is_ultra', axis=1)
target = df['is_ultra']
````

The data has the columns calls, minutes, messages, mb_used, and is_ultra. The is_ultra column is binary, where 1 indicates an Ultra plan and 0 indicates a Smart plan. The is_ultra column will therefore be our target, because the item we are classifying is to pick the right plan. This column will be dropped from the features, and the rest of the columns will remain in the feature set to train and test the model. There are 3214 observations.

````python
# train validation test 
features_temp, features_test, target_temp, target_test = train_test_split(
    features, target, test_size=0.25, random_state=12345)

features_train, features_valid, target_train, target_valid = train_test_split(
    features_temp, target_temp, test_size=0.3333, random_state=12345)

# hyperparameter-EXPLAINED BELOW
results = []

# random forest variations
for n in [10, 50, 100]:
    model = RandomForestClassifier(n_estimators=n, random_state=12345)
    model.fit(features_train, target_train)
    predictions = model.predict(features_valid)
    acc = accuracy_score(target_valid, predictions)
    results.append(('RandomForest', f'n_estimators={n}', acc))

# decision tree, (max_depth=None tells the model to not limit the depth of the tree)

#That means the tree can grow as deep as it wants to fit the training data.

#It’s a valid hyperparameter setting — not a mistake. 
for depth in [5, 10, None]:
    model = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model.fit(features_train, target_train)
    predictions = model.predict(features_valid)
    acc = accuracy_score(target_valid, predictions)
    results.append(('DecisionTree', f'max_depth={depth}', acc))

# logistic regression
model = LogisticRegression(random_state=12345, max_iter=1000)
model.fit(features_train, target_train)
predictions = model.predict(features_valid)
acc = accuracy_score(target_valid, predictions)
results.append(('LogisticRegression', 'default', acc))

#result
print("Model Comparison:")
for item in results:
    print(item[0], item[1], "Accuracy:", round(item[2], 3))


best_model = RandomForestClassifier(n_estimators=100, random_state=12345)
best_model.fit(features_train, target_train)
test_predictions = best_model.predict(features_test)
test_accuracy = accuracy_score(target_test, test_predictions)
print(f"Test set accuracy (RandomForest, n=100): {test_accuracy:.3f}")

#sanity check

most_common_class = target_train.mode()[0]  
baseline_predictions = [most_common_class] * len(target_valid)


baseline_accuracy = accuracy_score(target_valid, baseline_predictions)
print("Sanity check — baseline accuracy:", round(baseline_accuracy, 3))
````
